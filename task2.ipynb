```python
# Problem 1: Unique Words in a Sentence

sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary = 0
highest_employee = ""
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)

```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]
present_count = attendance.count("P")
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

text = "python"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

```python
# Problem 8: Convert List to Tuple

numbers = [10, 20, 30]
numbers_tuple = tuple(numbers)

print("Tuple:", numbers_tuple)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000
}
key = "Ravi"
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 80, 65, 75]
average = sum(marks) / len(marks)

print("Average Marks:", average)
```
