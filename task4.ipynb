```python
# Problem 1: Mobile Recharge Validation
def recharge_validation():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        amount = int(input("Enter recharge amount: "))
        
        if amount < 50:
            print("Recharge amount must be at least ₹50")
        
        elif amount in valid_plans:
            print("Recharge Successful for ₹", amount)
            break
        
        else:
            print("Invalid Plan. Try again.")
recharge_validation()
```

```python
# Problem 2: Inventory Reorder Alert

def check_inventory(products):
    
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

products = {
    "Soap": 10,
    "Shampoo": 25,
    "Toothpaste": 8,
    "Oil": 20
}
check_inventory(products)
```

```python
# Problem 3: Student Result Processing
def check_result(marks):  
    total = 0
    for m in marks:
        total = total + m
    
    average = total / len(marks)
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
marks_list = [60, 45, 70, 50]
result = check_result(marks_list)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare():
      while True:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")
        
        fare = 50 + (12 * km)
        
        if peak.lower() == "yes":
            fare = fare + (fare * 0.25)
        
        print("Total Fare: ₹", fare)
        
        retry = input("Do you want to calculate again? (yes/no): ")
        
        if retry.lower() != "yes":
            break

calculate_fare()
```

```python
# Problem 5: Attendance Eligibility
def attendance_check(attendance_list):
    
    present_days = 0 
    for day in attendance_list:
        if day == "P":
            present_days += 1 
    percentage = (present_days / len(attendance_list)) * 100 
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = attendance_check(attendance)
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        
        if ch in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and @#$)"

# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
